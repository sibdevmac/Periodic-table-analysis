```python
numeric_cols = [
    'AtomicNumber','AtomicMass','Density','MeltingPoint','BoilingPoint',
    'AtomicRadius','Electronegativity','IonizationEnergy',
    'AbundanceCrust','AbundanceUniverse'
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

<b> RQ1 How do atomic properties change across the periodic table?

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='AtomicNumber', y='Electronegativity', hue='Block')
plt.title("Electronegativity vs Atomic Number")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='AtomicNumber', y='AtomicRadius', hue='Block')
plt.title("Atomic Radius Trend Across Elements")
plt.show()
```

```python
corr = df[['AtomicNumber','Electronegativity','AtomicRadius','IonizationEnergy']].corr()
corr
```

<b> RQ2 Which elements are most abundant in Earth's crust vs the universe?

```python
top_crust = df.sort_values(by='AbundanceCrust', ascending=False).head(10)
top_universe = df.sort_values(by='AbundanceUniverse', ascending=False).head(10)

print("Top Elements in Earth's Crust")
print(top_crust[['EnglishName','AbundanceCrust']])

print("\nTop Elements in Universe")
print(top_universe[['EnglishName','AbundanceUniverse']])
```

<b> RQ3 How do physical properties influence the state of matter?

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='MeltingPoint', y='BoilingPoint', hue='StandardState')
plt.title("Melting vs Boiling Point by State")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Density', y='MeltingPoint', hue='StandardState')
plt.title("Density vs Melting Point")
plt.show()
```

<b><i>Treemap analysis

```python
import plotly.express as px
```

```python
df = df.dropna(subset=['DiscoveredBy'])
df['Label'] = df['EnglishName'] + " (" + df['Discovery'].astype(str) + ")"
fig = px.treemap(
    df,
    path=['DiscoveredBy', 'EnglishName'],
    values='AtomicNumber',
    hover_data=['AtomicNumber', 'Discovery'],
    title="Treemap of Elements by Discoverer"
)
fig.show()
```
